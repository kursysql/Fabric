# Data Wrangler
- załadowanie do ramki danych przykładowego zestawu danych
- DataWrangler
    - grupowanie wg kolumn dataSubType, dataSubType, category
    - zastosowanie funkcji count 
    - sortowanie wg liczby zdarzeń malejąco
    - dataType wielkimi literami
    - zmiana nazw kolumn: dataType -> Type, dataSubtype -> SubType
    - wygenerowanie kodu PySpark

```python
df = spark.read.format("csv").option("header","true").load("Files/sample_datasets/city_safety_seattle.csv")
# df now is a Spark DataFrame containing CSV data from "Files/sample_datasets/city_safety_seattle.csv".
display(df)
```

```python
%%sql
SELECT dataType, dataSubtype, category, count(*) AS cnt
FROM city_safety_seattle
GROUP BY ALL
ORDER BY cnt DESC
LIMIT 5
```

```python
# Code generated by Data Wrangler for PySpark DataFrame

from pyspark.sql import functions as F

def clean_data(df):
    # Performed 1 aggregation grouped on columns: 'dataType', 'dataSubtype', 'category'
    df = df.groupBy('dataType', 'dataSubtype', 'category').agg(F.count('dateTime').alias('dateTime_count'))
    df = df.dropna()
    df = df.sort(df['dataType'].asc(), df['dataSubtype'].asc(), df['category'].asc())
    # Sort by column: 'dateTime_count' (descending)
    df = df.sort(df['dateTime_count'].desc())
    # Convert text to uppercase in column: 'dataType'
    df = df.withColumn('dataType', F.upper(F.col('dataType')))
    # Rename column 'dataType' to 'Type'
    df = df.withColumnRenamed('dataType', 'Type')
    # Rename column 'dataSubtype' to 'Subtype'
    df = df.withColumnRenamed('dataSubtype', 'Subtype')
    return df

df_clean = clean_data(df)
display(df_clean)
```
