```sql
DROP VIEW IF EXISTS vw_Weather;
GO

CREATE VIEW vw_Weather AS 
SELECT d.Date, d.DayName, g.County, g.City, w.PrecipitationInches, w.AvgTemperatureFahrenheit
FROM Weather AS w 
JOIN Geography AS g ON w.GeographyID = g.GeographyID
JOIN Date AS d ON d.DateID = w.DateID
```

```sql
SELECT TOP 3 * FROM vw_Weather ORDER BY AvgTemperatureFahrenheit DESC
```
