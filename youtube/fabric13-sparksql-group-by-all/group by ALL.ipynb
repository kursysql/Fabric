# SparkSQL - group by all
- utworzenie warehouse SampleLH
- SparkSQL
    - tradycyjne grupowanie
    - GROUP BY ALL
- T-SQL
    - tradycyjne grupowanie
    - GROUP BY ALL (błąd składniowy)
- [demo na YouTube](https://youtu.be/ERoEmMo-yxU)
- [pozostałe filmy na kanale YT](https://www.youtube.com/watch?v=9FXwqwAcTYM&list=PLLbGIlLHMPz1hJpR84BWLpBSgRay10XD1)

```python
%%sql
SELECT * FROM publicholidays LIMIT 10
```

```python
%%sql
-- "tradycyjne" grupowanie
SELECT countryOrRegion, countryRegionCode, YEAR(date), count(*)
FROM publicholidays
WHERE date between '2010-01-01' AND '2024-12-31' AND countryRegionCode = 'US'
GROUP BY countryOrRegion, countryRegionCode, YEAR(date)
ORDER BY countryOrRegion, YEAR(date)
```

```python
%%sql
-- przykład z group by all
SELECT countryOrRegion, countryRegionCode, YEAR(date), count(*)
FROM publicholidays
WHERE date between '2010-01-01' AND '2024-12-31' AND countryRegionCode = 'US'
GROUP BY ALL
ORDER BY countryOrRegion, YEAR(date)
```

```python
-- T-SQL
SELECT countryOrRegion, countryRegionCode, YEAR(date), count(*)
FROM SampleLH.dbo.publicholidays
WHERE date between '2010-01-01' AND '2024-12-31' AND countryRegionCode = 'US'
GROUP BY countryOrRegion, countryRegionCode, YEAR(date)
ORDER BY countryOrRegion, YEAR(date)
```

```python
-- T-SQL, group by all - błąd składniowy
SELECT countryOrRegion, countryRegionCode, YEAR(date), count(*)
FROM SampleLH.dbo.publicholidays
WHERE date between '2010-01-01' AND '2024-12-31' AND countryRegionCode = 'US'
GROUP BY ALL
ORDER BY countryOrRegion, YEAR(date)
```
