## Odczyt pliku

```python
df = spark.read.format("csv").option("header","true").load("Files/Bikes/Mountain Bikes.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Bikes/Mountain Bikes.csv".
display(df.head(5))
```

## Pierwsze funkcje

```python
# wyświetlanie 5 wierszy za pomocą show
df.show(5)
```

```python
# show - skracanie łańcucha znaków
df.show(5, truncate=5)
```

```python
# show - wyłączenie skracania łańcuchów znaków
df.show(5, truncate=False)
```

```python
# sortowanie za pomocą orderBy (alias sort)
df.orderBy(df.ListPrice).show(5)
```

```python
# sortowanie malejące
df.orderBy(df.ListPrice.desc()).show(5)
```

```python
# sort jako alias orderBy
df.sort(df.ListPrice.desc()).show(5)
```

```python
# wyświetlenie schematu
df.printSchema()
```

## Zapis do tabeli delta

```python
# utworzenie nowej ramki (nie wyświetla żadnych danych w wyniku)
df_sorted = df.orderBy(df.Name)
```

```python
# sprawdzenie zawartości nowej ramki
df_sorted.show()
```

```python
# zapis zawartości ramki jako tabeli
df.write.format("delta").saveAsTable("products_from_file")
```

## Odczyt tabeli
